```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
legends = ["float", "half", "half2"]
colors = ["r", "g", "b"]
labels = ["1", "2", "4", "8"]

x = np.array([1, 2, 4, 8])
# perfFLT = [294.39, 153.21, 80.80, 44.33]
# perfHLF = [218.65, 114.45, 60.62, 32.84]
# perfHLF2= [166.29,  87.33, 46.84, 25.92]
# y1 = np.array([294.39, 153.21, 80.80, 44.33])
# y2 = np.array([218.65, 114.45, 60.62, 32.84])
# y3 = np.array([166.29,  87.33, 46.84, 25.92])

perfFLT = [298.55, 154.66, 81.09, 43.47]
perfHLF = [233.45, 121.76, 63.37, 33.39]
perfHLF2= [166.13,  87.76, 45.90, 25.32]
y1 = np.array([298.55, 154.66, 81.09, 43.47])
y2 = np.array([233.45, 121.76, 63.37, 33.39])
y3 = np.array([166.13,  87.76, 45.90, 25.32])
plt.figure()
plt.plot(x, 298.55 / y1, label=legends[0])
plt.plot(x, 298.55 / y2, label=legends[1])
plt.plot(x, 298.55 / y3, label=legends[2])
plt.legend()
# plt.scatter(x, 294.39 / y)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 5), dpi=500)
x = np.arange(len(labels))
width = 0.25
ax1.bar(x - width,  perfFLT, width=width, label=legends[0], color=colors[0])
ax1.bar(        x,  perfHLF, width=width, label=legends[1], color=colors[1])
ax1.bar(x + width, perfHLF2, width=width, label=legends[2], color=colors[2])

ax1.set_xlabel("nGPU", size=14)
ax1.set_ylabel("Elapsed time(s)", size=14)
ax1.set_xticks(x, labels)
ax1.grid(axis='y', linestyle='--', alpha=0.7)
# ax1.legend(fontsize="medium")
# ax.grid()

ax2.bar(x - width, [perfFLT[i] /  perfFLT[i] for i in range(len(perfFLT))], width=width, color=colors[0])
ax2.bar(        x, [perfFLT[i] /  perfHLF[i] for i in range(len(perfFLT))], width=width, color=colors[1])
ax2.bar(x + width, [perfFLT[i] / perfHLF2[i] for i in range(len(perfFLT))], width=width, color=colors[2])

ax2.set_xlabel("nGPU", size=14)
ax2.set_ylabel("Speedup", size=14)
ax2.set_xticks(x, labels)
ax2.grid(axis='y', linestyle='--', alpha=0.7)
# ax2.legend(fontsize="medium")

fig.legend(fontsize="medium", loc="upper right", bbox_to_anchor=(1.1, 0.935))
fig.tight_layout(pad=2.0)
fig.suptitle("Elapsed time and speedup for multiple GPUs", size=15)
fig.subplots_adjust(top=0.92)

plt.savefig("imagesDir/multigpu.pdf", dpi=500, bbox_inches = "tight")
```
